```python
'''
Author: Marco Cerignoni
Filename: M03 Lab.ipynb
Program Description: The program defines two classes, Vehicle and Automobile. It asks the user for information about a car (year, make, model, doors, roof type), creates an Automobile object with that data, and prints the carâ€™s details.'


Variable	    Type	Meaning
vehicle_type	str	    Describes the general type of vehicle (e.g., "car")
make	        str	    Manufacturer of the automobile (e.g., Ford, Toyota)
model	        str	    Specific model name (e.g., Mustang, Camry)
year	        int	    Year the vehicle was made
doors	        int	    Number of doors (must be 2 or 4)
roof	        str	    Type of roof ("sunroof" or "solid")
'''


class Vehicle:
    def __init__(self, vehicle_type: str) -> None:
        self.vehicle_type: str = vehicle_type
 

class Automobile(Vehicle):    
    def __init__(self, vehicle_type: str, year: int, make: str, model: str, doors: int, roof: str) -> None:
        super().__init__(vehicle_type)
        self.make: str = make
        self.model: str = model
        self.year: int = year

        if doors == 2 or doors == 4:
            self.doors: int = doors
        else:
            raise ValueError("Doors must be either 2 or 4")
    
        if roof.lower() == "sunroof" or roof.lower() == "solid":
            self.roof: str = roof
        else:
            raise ValueError("Roof must be either 'sunroof' or 'solid'")
    
    def __str__(self):
        return (
            f"Vehicle Type: {self.vehicle_type}\n"
            f"Year: {self.year}\n"
            f"Make: {self.make}\n"
            f"Model: {self.model}\n"
            f"Doors: {self.doors}\n"
            f"Roof Type: {self.roof}"
        )
    

year = int(input("Enter the year of the vehicle: "))
make = input("Enter the make of the vehicle: ")
model = input("Enter the model of the vehicle: ")  
doors = int(input("Enter the number of doors (2 or 4): "))
roof = input("Enter the roof type (sunroof or solid): ")
car = Automobile('car', year, make, model, doors, roof)

print(car)
```
