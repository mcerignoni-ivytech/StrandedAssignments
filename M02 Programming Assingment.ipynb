```python
#4.1
secret = 7
guess = int(input("Choose a number between 1 and 10: "))

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

```python
#4.2
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

```python
#6.2
guess_me = 7
number = 1

while(True):
    if guess_me == number:
        print("found it!")
        break
    elif guess_me < number:
        print("oops")
        break
    else:
        print("too low")
        number += 1
```
