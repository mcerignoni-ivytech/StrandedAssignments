```python
#13.1
from datetime import date

today = date.today()

today_string = today.isoformat()

with open("today.txt", "w") as file:
    file.write(today_string)

print("Date written to file:", today_string)
```

```python
#13.2
with open("today.txt", "r") as file:
    today_string = file.read()

print("Date read from file:", today_string)
```

```python
#13.3
from datetime import datetime

parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()

print("Parsed date:", parsed_date)
print("Type:", type(parsed_date))
```

```python
#15.1
import multiprocessing
import random
import time
from datetime import datetime


def wait_and_print():
    delay = random.random()

    time.sleep(delay)

    print("Waited", round(delay, 3), "seconds.")
    print("Current time is:", datetime.now())
    print("Process finished.\n")


if __name__ == "__main__":
    process1 = multiprocessing.Process(target=wait_and_print)
    process2 = multiprocessing.Process(target=wait_and_print)
    process3 = multiprocessing.Process(target=wait_and_print)

    process1.start()
    process2.start()
    process3.start()
    
    process1.join()
    process2.join()
    process3.join()

    print("All processes are done.")
```
